```python
# Import dependencies
import pandas as pd
```

```python
# Clean QBs
xl1 = pd.read_csv('data_raw/weekly_qb.csv')

xl1 = xl1.drop(columns=['Unnamed: 0', 'Unnamed: 2_level_0_Pts*'])

xl1 = xl1.rename(columns={'Unnamed: 0_level_0_Player':'player', 'Unnamed: 1_level_0_Game':'opponent',
       'Passing_Att':'passA', 'Passing_Cmp':"passC", 'Passing_Yds':"passY",
       'Passing_TD':'passT', 'Passing_Int':'passI', 'Passing_2Pt':'pass2', 'Rushing_Att':'rushA',
       'Rushing_Yds':'rushY', 'Rushing_TD':'rushT', 'Rushing_2Pt':'rush2', 'Receiving_Rec':'recC',
       'Receiving_Yds':'recY', 'Receiving_TD':'recT', 'Receiving_2Pt':'rec2', 'Fumbles_FL':'fum', 'Fumbles_TD':'fumT'
       })
xl1 = xl1[['player', 'season', 'week', 'team', 'opponent', 'passA', 'passC', 'passY', 'passT', 'passI', 'pass2',
       'rushA', 'rushY', 'rushT', 'rush2', 'recC', 'recY', 'recT', 'rec2',
       'fum']]
xl1[['tm1', 'tm2']] = xl1['opponent'].str.split('@', expand=True)

def teamPicker(row):
    team = row['team']
    if row['tm1']==team:
        opp = row['tm2']
    else:
        opp = row['tm1']
    return opp
xl1['opponent'] = xl1.apply(teamPicker, axis=1)

xl1 = xl1.drop(columns=['tm1', 'tm2'])

xl1[['player', 'todrop']] = xl1['player'].str.split('.\xa0', expand=True)
xl1 = xl1.drop(columns=['todrop'])

def dropLast(name):
    name = name[:-1]
    return name
xl1['player'] = xl1['player'].apply(dropLast)

xl1['pos'] = 'QB'

xl1['posRank'] = 'NA'
for i in range(len(xl1)):
    team = xl1.iloc[i]['team']
    season = xl1.iloc[i]['season']
    week = xl1.iloc[i]['week']
    player = xl1.iloc[i]['player']
    slicedf = xl1.loc[(xl1['team']==team) & (xl1['season']==season) & (xl1['week']==week)]
    slicedf = slicedf.sort_values(by=['passA'], ascending=False, ignore_index=True)
    posRank = slicedf.index[slicedf['player'] == player].to_list()[0]
    posRank += 1
    posRank = str(posRank)
    xl1.at[i, 'posRank'] = posRank
xl1['posRank'] = xl1['pos'] + xl1['posRank']
```
