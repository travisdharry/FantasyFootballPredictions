```python
import pandas as pd
```

```python
# Read csv
xl_rush = pd.read_csv("data_raw/Game_Logs_Runningback.csv")
xl_rec = pd.read_csv("data_raw/Game_Logs_Wide_Receiver_and_Tight_End.csv")
xl_pass = pd.read_csv("data_raw/Game_Logs_Quarterback.csv")
```

```python
#Select columns
rush_filtered = xl_rush[['Player Id', 'Year', 'Season', 'Week', 'Opponent',
       'Rushing Attempts', 'Rushing Yards',
       'Rushing TDs', 'Receptions', 'Receiving Yards',
       'Receiving TDs', 'Fumbles'
       ]]
rush_filtered = rush_filtered.rename(columns={'Player Id':'playerID', 'Year':'year', 'Week':'week', 'Opponent':'team', 'Season':'season', 
                'Rushing Attempts':'rushA', 'Rushing TDs':'rushTD', 'Rushing Yards':'rushY',
                'Receptions':'recA', 'Receiving Yards':'recY', 'Receiving TDs':'recTD', 
                'Fumbles':'fumbles'
                })
# Select only regular season
rush_filtered = rush_filtered[rush_filtered['season']=='Regular Season']
rush_filtered.drop(columns='season', inplace=True)
# Clean data
columnList = ['rushA', 'rushTD', 'rushY', 'recA', 'recTD', 'recY', 'fumbles']
for columnName in columnList:
       # Remove null values
       rush_filtered = rush_filtered[~rush_filtered[columnName].str.contains("--")]
       # Clean commas from numbers
       rush_filtered[columnName] = rush_filtered[columnName].str.replace(',', '').astype(int)
rush_filtered.tail(10)

```

```python
rec_filtered = xl_rec[['Player Id', 'Year', 'Season', 'Week', 'Opponent',
       'Rushing Attempts', 'Rushing Yards',
       'Rushing TDs', 'Receptions', 'Receiving Yards',
       'Receiving TDs', 'Fumbles'
       ]]
rec_filtered = rec_filtered.rename(columns={'Player Id':'playerID', 'Year':'year', 'Week':'week', 'Opponent':'team', 'Season':'season', 
                'Rushing Attempts':'rushA', 'Rushing TDs':'rushTD', 'Rushing Yards':'rushY',
                'Receptions':'recA', 'Receiving Yards':'recY', 'Receiving TDs':'recTD', 
                'Fumbles':'fumbles'
                })
```

```python
pass_filtered = xl_pass[['Player Id', 'Year', 'Season', 'Week', 'Opponent',
       'Rushing Attempts', 'Rushing Yards', 'Rushing TDs', 
       'Passes Attempted', 'Passing Yards', 'TD Passes', 
       'Ints', 'Fumbles'
       ]]
pass_filtered = pass_filtered.rename(columns={'Player Id':'playerID', 'Year':'year', 'Week':'week', 'Opponent':'team', 'Season':'season', 
                'Rushing Attempts':'rushA', 'Rushing TDs':'rushTD', 'Rushing Yards':'rushY',
                'Passes Attempted':'passA', 'Passing Yards':'passY', 'TD Passes':'passTD', 
                'Ints':'ints', 'Fumbles':'fumbles'
                })
```
