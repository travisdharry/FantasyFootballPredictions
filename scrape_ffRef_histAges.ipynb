```python
# Import Library
import pandas as pd
import re

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium.webdriver.support.ui import WebDriverWait

from selenium.common.exceptions import TimeoutException
```

```python
# Set Selenium settings
capa = DesiredCapabilities.CHROME
capa["pageLoadStrategy"] = "none"

# Create empty df
df = pd.DataFrame(columns=['Player', 'Age', 'Season'])
```

```python
# Scrape web for stats
for role in ['passing', 'receiving']:
    for season in range(2010, 2022):

        url = f"https://www.pro-football-reference.com/years/{season}/{role}.htm"

        PATH = "/Applications/chromedriver"
        driver = webdriver.Chrome(service=Service(PATH), desired_capabilities=capa)
        wait = WebDriverWait(driver, 20)
        driver.get(url)

        wait.until(EC.presence_of_element_located((By.XPATH, f"//table[@id='{role}']")))
        driver.execute_script("window.stop();")

        result = pd.read_html(driver.find_element(By.XPATH, value=f"//table[@id='{role}']").get_attribute("outerHTML"))
        result = result[0]
        result = result[['Player', 'Age']]
        result['Season'] = season

        df = pd.concat([df, result], axis=0, ignore_index=True)
df
```

```python
# Scrape web for stats
for role in ['rushing', 'kicking']:
    for season in range(2010, 2022):

        url = f"https://www.pro-football-reference.com/years/{season}/{role}.htm"

        PATH = "/Applications/chromedriver"
        driver = webdriver.Chrome(service=Service(PATH), desired_capabilities=capa)
        wait = WebDriverWait(driver, 20)
        driver.get(url)

        wait.until(EC.presence_of_element_located((By.XPATH, f"//table[@id='{role}']")))
        driver.execute_script("window.stop();")

        result = pd.read_html(driver.find_element(By.XPATH, value=f"//table[@id='{role}']").get_attribute("outerHTML"))
        result = result[0]
        result.columns = result.columns.get_level_values(1)
        result = result[['Player', 'Age']]
        result['Season'] = season

        df = pd.concat([df, result], axis=0, ignore_index=True)
df
```

```python
df['Player'] = df['Player'].str.replace("*", "")
df['Player'] = df['Player'].str.replace("+", "")
df.head(50)
```

```python
df = df.loc[df['Age']!='Age']
df.reset_index(inplace=True, drop=True)
df.head(50)
```
